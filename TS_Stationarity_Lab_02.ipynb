# Time Series Stationarity

```r
set.seed(1)  # for replication
y <- ts(rnorm(500))# white noise from normal dist.  WN(mu,sigma)
```

```r
summary(y)
```

```r
plot(y)
```

```r
# autocovariance

acf(y,type="covariance", plot=F)
```

```r
acf(y)
```

```r
# using poison for counting process
set.seed(1)
y<- ts(rpois(5000,3))
```

```r
plot(y)
```

```r
acf(y)
```

## Simulating Random Walk model

set seed

2- start from 0

3- loop over a sample of 1000. Every step, add a white noise to the previous value

4- convert to ts object

5- plot

6 take difference and then plot

```r
set.seed(2)
n=1000
x= rep(0,n)

for(i in 2:n)
    {
    x[i]=x[i-1] + rnorm(1)
}
```

```r
head(x)
summary(x)
```

```r
xt= ts(x)
plot(xt)
```

```r
acf(xt)
```

X_t-X_(t-1)

```r
dxt=diff(xt)# taking the difference operator on the data
plot(dxt)
```

```r
acf(dxt)
```
